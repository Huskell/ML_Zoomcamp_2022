```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
```

```python
df2 = df.loc[:, ['latitude',
                'longitude',
                'housing_median_age',
                'total_rooms',
                'total_bedrooms',
                'population',
                'households',
                'median_income',
                'median_house_value',
                'ocean_proximity']]
```

```python
df2.loc[:, ['rooms_per_household']] = df2['total_rooms'] / df2['households']
df2.loc[:, ['bedrooms_per_room']] = df2['total_bedrooms'] / df2['total_rooms']
df2.loc[:, ['population_per_household']] = df2['population'] / df2['households']
```

### Question 1

What is the most frequent observation (mode) for the column ocean_proximity?

```python
df2.loc[:,'ocean_proximity'].mode()
```
